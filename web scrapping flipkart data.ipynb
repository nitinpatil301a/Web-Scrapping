```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
req=requests.get(url)
soup=BeautifulSoup(req.content,'html.parser')
```

```python
containers=soup.find_all(class_="_1-2Iqu row")
name=[container.find(class_="_3wU53n").get_text() for container in containers]
rating=[container.find(class_="hGSR34").get_text() for container in containers]
price=[container.find(class_="_1vC4OE _2rQ-NK").get_text() for container in containers]
rating
```

```python
import pandas as pd
flipkart_iphone_data=pd.DataFrame({'Product_name':name,'Product_rating':rating,'Product_price':price})
flipkart_iphone_data
```
